## Heatmaps

Heatmaps in Seaborn are a powerful data visualization technique that uses color to represent different levels of data magnitude. The key points about Seaborn heatmaps are:

1. Heatmaps organize data in a grid, with different colors or shades indicating different levels of the data's magnitude. This allows for immediate recognition of patterns, such as clusters, trends, and anomalies.[1][2][3][4]

2. Heatmaps are best suited for datasets where you can represent values as colors, typically continuous or discrete numerical data. They are useful for visualizing the correlation matrix between multiple variables.[1][3][4]

3. Seaborn provides an easy-to-use heatmap function, `sns.heatmap()`, that allows you to create highly customized heatmap visualizations. You can adjust the color palette, handle missing data, scale the data, add annotations, and control the dimensions of the heatmap.[1][2][3][4][5]

4. Best practices for using Seaborn heatmaps include choosing the right color palette, handling missing data thoughtfully, properly scaling the data, using annotations sparingly, and adjusting the heatmap dimensions to ensure patterns are clearly visible.[1][3][4]

5. Seaborn heatmaps are commonly used in data analysis, machine learning, and finance applications to quickly identify patterns and trends in large datasets.[4][5]

In summary, Seaborn heatmaps are a versatile and powerful data visualization tool that allow you to easily create customized, high-quality heatmap visualizations to explore and communicate insights from your data.

```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index = 'year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False) 
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5) #linewidth is used for varying the gap 
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d') #annot will help display the values in the box, fmt is for the type data in this case it is integer 
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='viridis') #cmap is used to change the colour 
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='summer') #cmap is used to change the colour 
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='winter') #cmap is used to change the colour 
```

## Clustermaps

A cluster map (or clustermap) in Seaborn is a data visualization technique that plots a hierarchically clustered heatmap of a rectangular dataset. The key points about Seaborn cluster maps are:

1. Cluster maps organize data in a grid, with different colors or shades indicating different levels of the data's magnitude. This allows for immediate recognition of patterns, such as clusters, trends, and anomalies in the data.[1][2][3][4]

2. The `sns.clustermap()` function in Seaborn is used to create these cluster map visualizations. It takes a rectangular dataset as input and performs hierarchical clustering on the rows and/or columns, then displays the resulting heatmap with the clustered data.[1][2][3][4][5]

3. Cluster maps are useful for visualizing the correlation matrix between multiple variables, as well as identifying groups or patterns in high-dimensional data.[1][3][4]

4. Seaborn provides many customization options for cluster maps, such as adjusting the color palette, handling missing data, scaling the data, adding annotations, and controlling the dimensions of the heatmap.[1][2][3][4][5]

5. Cluster maps are commonly used in data analysis, machine learning, and finance applications to quickly identify patterns and trends in large datasets.[4][5]

In summary, Seaborn cluster maps are a powerful data visualization tool that allow you to easily create customized, hierarchically clustered heatmaps to explore and communicate insights from your data.

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
 sns.clustermap(x,z_score=0,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,row_cluster=False,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,col_cluster=False,metric='correlation')
```
