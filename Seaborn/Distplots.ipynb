## Distplots

The `distplot` in Seaborn was a function designed to visualize the distribution of a univariate set of observations. It combined multiple functionalities from Seaborn and Matplotlib to create a comprehensive view of data distributions, including histograms, kernel density estimates (KDEs), and rug plots. The primary purpose of `distplot` was to provide a quick and easy way to visualize the distribution of a single variable and understand its underlying patterns, such as its central tendency, spread, skewness, and presence of outliers[1][2][4][5][6].

Key features of `distplot` included:

- **Histogram**: By default, `distplot` displayed a histogram, which is a bar plot that shows the frequency (or density) of occurrences for different ranges of values[2][4][6].
- **KDE**: Optionally, a kernel density estimate could be overlaid on the histogram to show a smooth estimate of the distribution. This was useful for identifying the shape of the distribution[2][4][6].
- **Rug Plot**: A rug plot could also be added, which placed small ticks along the x-axis to indicate the location of individual data points. This provided a sense of the data's density and distribution[2][4][6].
- **Fit**: `distplot` could fit and plot a parametric distribution to the observed data, providing a way to compare the empirical distribution of the data to theoretical models[5].

However, as of Seaborn version 0.11, `distplot` has been deprecated and is planned to be removed in a future release (v0.14.0). It has been replaced by two new functions: `histplot()` and `displot()`. These functions offer similar capabilities but with a modern API and enhanced features. `histplot()` is an axes-level function for plotting histograms, including with kernel density smoothing, while `displot()` is a figure-level function that provides access to several approaches for visualizing univariate or bivariate distributions, including histograms, KDEs, and empirical cumulative distribution functions (ECDFs)[1][3][5][7][8].

The deprecation of `distplot` and the introduction of `histplot` and `displot` reflect Seaborn's ongoing development and its aim to provide more flexible and powerful tools for statistical data visualization[1][3][5][7][8].

```python
import seaborn as sns 
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True) #Filling the null values
```

## Plotting a Distplot

```python
sns.distplot(titanic['age'])
```

## Plotting only Histogram

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.histplot(titanic['age'],bins=10,kde=False)
```

```python
sns.displot(titanic['age'],bins=10,kde=False)
```

## Plotting Only the KDE

```python
sns.distplot(titanic['age'],hist=False)
```

## Plotting Only the RugPlot

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
