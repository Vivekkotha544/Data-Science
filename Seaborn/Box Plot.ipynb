## Box Plot

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset that summarizes key statistics such as the median, quartiles, and outliers. It provides insights into the spread and central tendency of the data. Box plots are used for various types of data, especially for comparing multiple groups[1][2][3][7][8][9][10][11][12][13][14].

## Elements of a Box Plot

- **Minimum**: The smallest value in the dataset, excluding any outliers.
- **First Quartile (Q1)**: Also known as the lower quartile, it marks the 25th percentile of the data. 25% of the data points are below this value.
- **Median (Q2)**: The middle value of the dataset when it is ordered from least to greatest. It divides the dataset into two equal parts.
- **Third Quartile (Q3)**: Also known as the upper quartile, it marks the 75th percentile of the data. 75% of the data points are below this value.
- **Maximum**: The largest value in the dataset, excluding any outliers.
- **Interquartile Range (IQR)**: The difference between the third quartile (Q3) and the first quartile (Q1). It measures the spread of the middle 50% of the data.
- **Outliers**: Data points that fall outside of the typical range (usually defined as 1.5 times the IQR above Q3 or below Q1).
- **Whiskers**: Lines that extend from the box to the highest and lowest values within 1.5 times the IQR from the quartiles, excluding outliers.
- **Box**: The central rectangle spanning from Q1 to Q3.

## Interpretation

- The length of the box (IQR) indicates the variability within the central 50% of the data.
- The position of the median within the box shows the skewness of the data. If the median is closer to Q1 or Q3, it indicates skewness towards the lower or higher values, respectively.
- Outliers are indicated as individual points beyond the whiskers, highlighting data points that differ significantly from the rest of the dataset.
- Comparing box plots across groups can reveal differences in medians, variability, and the presence of outliers among the groups.

## Uses

Box plots are particularly useful for:
- Comparing distributions between different groups or categories.
- Identifying outliers that may warrant further investigation.
- Visualizing the spread and skewness of the data.
- Providing a concise summary of the data's distribution without making any assumptions about the underlying statistical distribution[1][2][3][7][8][9][10][11][12][13][14].

In summary, box plots serve as a powerful tool for exploratory data analysis, allowing for quick visual comparisons and identification of data characteristics across different datasets or groups.

```python
import seaborn as sns 
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.catplot (x='day',y='total_bill',kind='box',data = tips)
```

```python
sns.catplot (x='day',y='total_bill',kind='box',hue='sex',data = tips)
```

## Violin Plot

A violin plot is a method for visualizing the distribution of quantitative data across different levels of one or more categorical variables.

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=2,data=tips) #using split
```
